```python
# Import necessary libraries
import pandas as pd

# Load the CSV file
file_path = 'googleplaystore.csv'  # Replace with the actual path to your file
data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
print("First few rows of the data:")
data.head()

# Step 1: Identify missing values
print("\nMissing values per column:")
print(data.isnull().sum())

# Step 2: Drop rows with missing values or handle them accordingly
# Option 1: Drop rows with any missing values
cleaned_data = data.dropna()

# Option 2: Fill missing values (e.g., for numerical columns)
# cleaned_data = data.fillna(data.median())

# Step 3: Remove duplicates (if any)
cleaned_data = cleaned_data.drop_duplicates()

# Step 4: Convert columns to appropriate data types (if necessary)
# For example, converting the "Installs" column from string to integer
# First, remove any non-numeric characters (like commas) from the 'Installs' column
cleaned_data['Installs'] = cleaned_data['Installs'].str.replace('[+,]', '', regex=True).astype(float)

# Step 5: Handle inconsistent data (e.g., cleaning 'Price' and 'Size' columns)
# Example: Remove the dollar sign from 'Price' column and convert to float
if 'Price' in cleaned_data.columns:
    cleaned_data['Price'] = cleaned_data['Price'].str.replace('$', '').astype(float)

# Step 6: Display cleaned data
print("\nCleaned data:")
cleaned_data.head()

# Step 7: Save the cleaned data to a new CSV file
cleaned_data.to_csv('cleaned_googleplaystore.csv', index=False)
print("\nCleaned data saved to 'cleaned_googleplaystore.csv'")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Distribution of app ratings
plt.figure(figsize=(10,5))
sns.histplot(cleaned_data['Rating'], bins=20, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# Example: Top 10 categories by number of apps
top_categories = cleaned_data['Category'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_categories.index, y=top_categories.values, palette='viridis')
plt.title('Top 10 Categories by Number of Apps')
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by 'App' and sum the 'Installs' column to find the total installs for each app
most_installed_apps = cleaned_data.groupby('App')['Installs'].sum().sort_values(ascending=False)

# Display the app with the highest installs
print("App with the most installs:")
print(most_installed_apps.head(1))
```

```python
# Total installs by category
installs_by_category = cleaned_data.groupby('Category')['Installs'].sum().sort_values(ascending=False)
print("\nCategories with the most installs:")
print(installs_by_category.head(10))

# Average rating by category
avg_rating_by_category = cleaned_data.groupby('Category')['Rating'].mean().sort_values(ascending=False)
print("\nAverage rating by category:")
print(avg_rating_by_category.head(10))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file (replace with the correct path)
file_path = 'googleplaystore.csv'  # Update this path
data = pd.read_csv(file_path)

# Step 1: Clean the 'Installs' column
data['Installs'] = data['Installs'].astype(str)
data['Installs'] = data['Installs'].str.replace('[+,]', '', regex=True)
data['Installs'] = pd.to_numeric(data['Installs'], errors='coerce')

# Step 2: Clean the 'Price' column
if 'Price' in data.columns:
    data['Price'] = data['Price'].astype(str)
    data['Price'] = data['Price'].str.replace('$', '')
    data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Step 3: Clean the 'Size' column
if 'Size' in data.columns:
    data['Size'] = data['Size'].replace('Varies with device', np.nan)
    data['Size'] = data['Size'].str.replace('M', '')
    data['Size'] = data['Size'].str.replace('k', '')
    data['Size'] = pd.to_numeric(data['Size'], errors='coerce')

# Step 4: Clean the 'Android Ver' column
if 'Android Ver' in data.columns:
    data['Android Ver'] = data['Android Ver'].replace('Varies with device', np.nan)
    data['Android Ver'] = data['Android Ver'].str.extract(r'(\d+\.\d+)')  # Extract only numeric versions
    data['Android Ver'] = pd.to_numeric(data['Android Ver'], errors='coerce')  # Convert to float

# Step 5: Filter apps that require Android versions below 4.0
apps_below_4 = data[data['Android Ver'] < 4.0]

# Step 6: Insights
# Total number of apps below Android version 4.0
total_apps_below_4 = apps_below_4.shape[0]
print(f"Total number of apps requiring Android versions below 4.0: {total_apps_below_4}")

# Average rating of apps below Android version 4.0
average_rating_below_4 = apps_below_4['Rating'].mean()
print(f"Average rating of apps requiring Android versions below 4.0: {average_rating_below_4:.2f}")

# Total installs for apps below Android version 4.0
total_installs_below_4 = apps_below_4['Installs'].sum()
print(f"Total installs for apps requiring Android versions below 4.0: {total_installs_below_4}")

# Top 5 most installed apps below Android version 4.0
top_10_installed_below_4 = apps_below_4[['App', 'Installs']].sort_values(by='Installs', ascending=False).head(10)
print("\nTop 10 most installed apps requiring Android versions below 4.0:")
print(top_10_installed_below_4)
```

```python
# Step 5: Clean and convert the 'Last Updated' column
if 'Last Updated' in data.columns:
    data['Last Updated'] = pd.to_datetime(data['Last Updated'], errors='coerce')

# Step 6: Filter apps that require Android versions below 4.0
apps_below_4 = data[data['Android Ver'] < 4.0]

# Step 7: Top 10 most installed apps below Android version 4.0
top_10_installed_below_4 = apps_below_4[['App', 'Installs', 'Last Updated']].sort_values(by='Installs', ascending=False).head(10)

# Step 8: Format the 'Last Updated' date to dd/mm/yyyy
top_10_installed_below_4['Last Updated'] = top_10_installed_below_4['Last Updated'].dt.strftime('%d/%m/%Y')

# Print the top 10 most installed apps and their last updated dates
print("Top 10 most installed apps requiring Android versions below 4.0:")
print(top_10_installed_below_4)

# Step 9: Visualize the top 10 most installed apps
plt.figure(figsize=(12, 6))
sns.barplot(x='Installs', y='App', data=top_10_installed_below_4, palette='viridis')
plt.title('Top 10 Most Installed Apps Requiring Android Versions Below 4.0')
plt.xlabel('Number of Installs')
plt.ylabel('App Name')

# Annotate with the last updated date
for index, row in top_10_installed_below_4.iterrows():
    plt.text(row['Installs'], index, row['Last Updated'], color='black')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'googleplaystore.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Step 1: Filter Health Apps
health_apps = data[data['Category'].str.contains('HEALTH', case=False, na=False)]

# Step 2: Clean the Installs column (remove commas, plus signs and convert to numeric)
health_apps['Installs'] = health_apps['Installs'].str.replace(',', '').str.replace('+', '').astype(float)

# Step 3: Convert 'Last Updated' to datetime format and extract the year
health_apps['Last Updated'] = pd.to_datetime(health_apps['Last Updated'], errors='coerce')
health_apps['Year'] = health_apps['Last Updated'].dt.year

# Step 4: Aggregate total installs per year
downloads_by_year = health_apps.groupby('Year')['Installs'].sum().reset_index()

# Step 5: Visualize the results
plt.figure(figsize=(10, 6))
plt.bar(downloads_by_year['Year'], downloads_by_year['Installs'], color='skyblue')
plt.title('Total Downloads of Health Apps by Year', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Installs', fontsize=12)
plt.xticks(downloads_by_year['Year'].astype(int), rotation=45)
plt.tight_layout()

# Show the plot
plt.show()

# Display the summary table
print(downloads_by_year)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
file_path_store = 'googleplaystore.csv'  # Replace with your file path for googleplaystore.csv
file_path_reviews = 'googleplaystore_user_reviews.csv'  # Replace with your file path for googleplaystore_user_reviews.csv

# Read the data
store_data = pd.read_csv(file_path_store)
reviews_data = pd.read_csv(file_path_reviews)

# Step 1: Clean the reviews data by removing duplicates (optional, depending on dataset)
reviews_data = reviews_data.drop_duplicates(subset=['App'])

# Step 2: Merge the datasets on the 'App' column (inner join to keep only apps with reviews)
merged_data_with_reviews = pd.merge(store_data, reviews_data[['App']], on='App', how='inner')

# Step 3: Count how many reviews each app has (only for apps with reviews)
apps_with_reviews_count = merged_data_with_reviews['App'].value_counts()

# Step 4: Select the top 5 apps with the most reviews for the pie chart
top_5_apps_with_reviews = apps_with_reviews_count.head(5)

# Step 5: Plot the pie chart for the top 5 apps with reviews
plt.figure(figsize=(8, 8))
plt.pie(top_5_apps_with_reviews.values, labels=top_5_apps_with_reviews.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of Reviews for Top 5 Apps', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the pie chart
plt.show()

# Display the top 5 apps with most reviews
print("Top 5 Apps with Reviews:")
print(top_5_apps_with_reviews)
```
